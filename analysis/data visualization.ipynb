```python
##plotting net revenue and spending for android platforms over time
spendings.head()

spendings_android = spendings.query('platform=="android"')
androidspendings = spendings_android.groupby(['month'])['spendings'].agg('sum')

androidspendings = androidspendings.to_frame().reset_index()
androidspendings.columns = ['month', 'spendings']


spendings_ios = spendings.query('platform=="ios"')
iosspendings = spendings_ios.groupby(['month'])['spendings'].agg('sum')

iosspendings = iosspendings.to_frame().reset_index()
iosspendings.columns = ['month', 'spendings']


revenue_android = subscriptions.query('signup_platform=="android"')
androidrevenue = revenue_android.groupby(['month'])['net_revenue'].agg('sum')

androidrevenue = androidrevenue.to_frame().reset_index()
androidrevenuecolumns = ['month', 'revenue']


revenue_ios = subscriptions.query('signup_platform=="ios"')
iosrevenue = revenue_ios.groupby(['month'])['net_revenue'].agg('sum')

iosrevenue = iosrevenue.to_frame().reset_index()
iosrevenuecolumns = ['month', 'revenue']


android = pd.merge(androidspendings, androidrevenue, on = 'month')
ios = pd.merge(iosspendings, iosrevenue, on = 'month')


android['month'].replace(10, 'Oct 2016',inplace=True)
android['month'].replace(11, 'Nov 2016',inplace=True)
android['month'].replace(12, 'Dec 2016',inplace=True)
android['month'].replace(1, 'Jan 2017',inplace=True)

android = android.reindex([1,2,3,0])

ios['month'].replace(10, 'Oct 2016',inplace=True)
ios['month'].replace(11, 'Nov 2016',inplace=True)
ios['month'].replace(12, 'Dec 2016',inplace=True)
ios['month'].replace(1, 'Jan 2017',inplace=True)

ios = ios.reindex([1,2,3,0])

android['profit'] = android['net_revenue'] - android['spendings']

ios['profit'] = ios['net_revenue'] - ios['spendings']

```

```python
fig = plt.plot( 'month', 'spendings', data=android, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'month', 'net_revenue', data=android, marker='', color='olive', linewidth=2)
plt.legend()
plt.title('Android platforms - net revenue and spending over time')
plt.xlabel('Time period')
plt.ylabel('Amount in euros')

plt.show()

plt.savefig('androidtimeseries.jpg', bbox_inches='tight')   # save the figure to file


plt.close()
```

```python
fig = plt.plot( 'month', 'spendings', data=ios, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'month', 'net_revenue', data=ios, marker='', color='olive', linewidth=2)
plt.legend()
plt.title('IOS platforms - net revenue and spending over time')
plt.xlabel('Time period')
plt.ylabel('Amount in euros')

plt.show()

plt.savefig('iostimeseries.jpg', bbox_inches='tight')   # save the figure to file

plt.close()
```

```python
##plotting net revenue and spending for android platforms over time
spendings.head()

spendings_1 = spendings.query('country_code=="CH"')
spendings_1 = spendings_1.groupby(['month'])['spendings'].agg('sum')

spendings_1 = spendings_1.to_frame().reset_index()
spendings_1.columns = ['month', 'spendings']


revenue_1 = subscriptions.query('signup_country_code=="UK"')
revenue_1 = revenue_1.groupby(['month'])['net_revenue'].agg('sum')

revenue_1 = revenue_1.to_frame().reset_index()
revenue_1.columns = ['month', 'revenue']

mkchannel1 = pd.merge(revenue_1, spendings_1, on = 'month')

mkchannel1['month'].replace(10, 'Oct 2016',inplace=True)
mkchannel1['month'].replace(11, 'Nov 2016',inplace=True)
mkchannel1['month'].replace(12, 'Dec 2016',inplace=True)
mkchannel1['month'].replace(1, 'Jan 2017',inplace=True)

mkchannel1 = mkchannel1.reindex([1,2,3,0])

mkchannel1['profit'] = mkchannel1['revenue'] - mkchannel1['spendings']
```
